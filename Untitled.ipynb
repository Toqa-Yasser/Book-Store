```python
i=1
import csv
import bs4
from bs4 import BeautifulSoup
import csv
import requests
req=requests.get('https://books.toscrape.com/catalogue/category/books_1/index.html')
soup=BeautifulSoup(req.content,'html.parser')
categorys=soup.find('ul',attrs={'class':'nav nav-list'}).find('li').find('ul').find_all('li')
with open ('books.csv',mode='w',newline='') as f:
    writer=csv.writer(f)
    writer.writerow(['Catogery','title','price','rate'])
for category in categorys:
    i=i+1
    namecat=category.find('a').text.strip()
    l=category.find('a').get('href')[2:]
    linksite='https://books.toscrape.com/catalogue/category'
    link=linksite+l
    req=requests.get(link)
    soup=BeautifulSoup(req.content,'html.parser')
    try:
        numofpages=int(soup.find('section',attrs=None).find('div',attrs=None).find('li',attrs={'class':'current'}).text.strip()[10:])
    except:
        numofpages=1
    for pn in range(1,numofpages+1):
        if numofpages==1:
            req=requests.get(link)
        else:
            if len(namecat.split())==1:
                req=requests.get(f'https://books.toscrape.com/catalogue/category/books/{namecat.lower()}_{i}/page-{pn}.html')
            else:
                joinname ='-'.join(namecat.split()).lower()
                req=requests.get(f'https://books.toscrape.com/catalogue/category/books/{joinname}_{i}/page-{pn}.html')
        soup=BeautifulSoup(req.content,'html.parser')
        books=soup.find('section').find('div',attrs=None).find_all('li',attrs={'class':'col-xs-6 col-sm-4 col-md-3 col-lg-3'})
        for book in books:
            title=book.find('h3').find('a').attrs['title']
            stars=book.find('p').get('class')[1]
            price=float(book.find('div',attrs={'class':'product_price'}).find('p').text[1:])
            with open ('books.csv',mode='a',newline='') as f:
                writer=csv.writer(f)
                writer.writerow([namecat,title,price,stars])
```
